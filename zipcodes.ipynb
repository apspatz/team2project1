```python
#Importing Dependencies
import pandas as pd
import json
import requests
from pprint import pprint

#Importing API key for zipcodeapi
from config import zipcodeapi
```

```python
#Parameters for API
zip_code = "27560"
distance = "25"
units = "mile"

#Pulling zip code data
url = "https://www.zipcodeapi.com/rest/{0}/radius.json/{1}/{2}/{3}".format(zipcodeapi,zip_code,distance,units)
response = requests.get(url).json()

pprint(response)
```

```python
#Creating and formatting zipcode dataframe
zipdata = response["zip_codes"]
zipdata_pd = pd.DataFrame(zipdata)
zipdata_pd = zipdata_pd.sort_values("city")
zipdata_pd = zipdata_pd[{"zip_code","city"}]
zipdata_pd
```

```python
#Saving dataframe to CSV
zipdata_pd.to_csv("zipcodes.csv")
```
