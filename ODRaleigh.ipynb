```python
import requests
import json
from pprint import pprint
import pandas as pd

url = "https://services.arcgis.com/v400IkDOw1ad7Yad/arcgis/rest/services/Police_Incidents/FeatureServer/0/query?where=1%3D1&outFields=crime_description,city,district,reported_year,reported_month,crime_category&outSR=4326&f=json"

response = requests.get(url).json()
pprint(response)
```

```python
crimedata = response["features"]
formatted_data = [crime['attributes'] for crime in crimedata]
# pprint(formatted_data)

crimedata_pd = pd.DataFrame(formatted_data)
crimedata_pd.head(100)
```

```python
district_count = crimedata_pd['district'].value_counts()

cat_types = crimedata_pd['crime_category'].value_counts()
print(cat_types)
```

```python
crimedata_pd["crime_description"]=crimedata_pd["crime_description"].str.split("/",1,False)
crimedata_pd.head()
```

```python
cat = []
descrip =[]
for index, row in crimedata_pd.iterrows():
    descrip.append(row["crime_description"][1])
    cat.append(row["crime_description"][0])
crimedata_pd["description"]=descrip
crimedata_pd["crime_cat"]=cat
crimedata_pd.drop(columns="crime_description",inplace=True)
crimedata_pd.head()
```

```python
crimedata_pd["crime_cat"].value_counts()
```

```python
crimedata_pd['description'].value_counts()
```

```python
crimedata_pd['description'] = crimedata_pd['description'].replace({'Overdose':'Deceased Person','Suicide':'Deceased Person',
                                                                  'Overdose Death':'Deceased Person'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Missing Person (18 & over)':'Missing Person',
                                                                  'Runaway':'Missing Person', 'Missing Person (17 & under)': 'Missing Person'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Simple':'Assault','All Other':'Assault',
                                                                  'Aggravated':'Assault', 'Assault with an Object':'Assault'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Forcible Rape':'Sex crime', 'Forcible Fondling':'Sex crime',
                                                                  'Peeping Tom':'Sex crime','Forcible Sodomy':'Sex crime',
                                                                  'Statutory Rape':'Sex crime', 'Incest':'Sex crime'})
crimedata_pd['description'] = crimedata_pd['description'].replace({'Mental Commitment':'Other crimes','Found Property':'All Other Non-Offenses',
                                                                  'Lost Property':'Other crimes', 'Nonviolent (eg: neglect)':'Other crimes',
                                                                  'Injured Person':'Other crimes','Citizen Request':'All Other Non-Offenses',
                                                                  'Child Abuse or Neglect':'Other crimes','Truancy':'Other crimes'})
crimedata_pd['description'].value_counts()
```

```python
crimedata_pd = crimedata_pd.sort_values(['reported_year','reported_month'])
crimedata_pd = crimedata_pd.reset_index(drop=True)
crimedata_pd.head(50)
```

```python
crimes_pd = crimedata_pd.loc[crimedata_pd['description'] != 'All Other Non-Offenses']
crimes_pd = crimes_pd.reset_index(drop=True)
crimes_pd.head(50)
```

```python
crimes_grouped = crimes_pd.groupby('district')
crimes_grouped['description'].value_counts()
```
